```r
library(tidyverse, warn.conflicts=FALSE)
library("ggpubr")
```

## chose time point

```r
tp <- "diff"
#tp <- "1"
#tp <- "2" 
```

## Importing data: UKBiobank and emerge LV, phenotype traits strings, RNA projection data

```r
# Import data
gls_phenomexcan <- readRDS("data/input/gls-summary-phenomexcan.rds")
gls_emerge <- readRDS("data/input/gls-summary-emerge.rds")

# Type 1 diabetes traits in dataset 
UK_biobank_traits <-  c('Type 1 diabetes with ketoacidosis','Type 1 diabetes', 'Non-cancer illness code, self-reported: type 1 diabetes', 'Type 1 diabetes without complications','Type 1 diabetes with ophthalmic complications')
emerge_trait <- 'Type 1 diabetes'

# Daisy RNAseq projection
DAISY_diff_proj <- readRDS(paste0("output/gp_RNA_proj_",tp,".rds"))
```

## Filter LVs associated with T1D in phenomexcan and emerge at p-value <0.01

```r
#987 LVs in both cohort, narrow down to phenotypes related to T1D
phe_T1D_LVs <- gls_phenomexcan[gls_phenomexcan$phenotype_desc %in% UK_biobank_traits,]
e_T1D_LVs <- gls_emerge[gls_emerge$phenotype_desc %in% emerge_trait,]

# There are 17 LVs invovled in T1D from both phenomexcan and e-merge at a nominal p-value of 0.01
#122
phe_T1D_LVs_nom_0.01 <- phe_T1D_LVs[phe_T1D_LVs$pvalue < 0.01,]
colnames(phe_T1D_LVs_nom_0.01)
dim(phe_T1D_LVs_nom_0.01)
#19
e_T1D_LVs_nom_0.01 <- e_T1D_LVs[e_T1D_LVs$pvalue < 0.01, ]
#17
phe_U_e_nom_0.01 <- inner_join(phe_T1D_LVs_nom_0.01,e_T1D_LVs_nom_0.01, by = "lv")
wider <- phe_T1D_LVs_nom_0.01[,! colnames(phe_T1D_LVs_nom_0.01) == "phenotype_desc"]%>%
  pivot_wider(
    names_from =phenotype,
      names_sep = "_",
      values_from = c('pvalue','fdr'),
  )

top_17_LV <- inner_join(wider,e_T1D_LVs_nom_0.01, by = "lv")
neg_log10 <- function(x, na.rm = FALSE) -1*log10(x)
top_17_LV_log <- top_17_LV %>% mutate_if(is.double,neg_log10)
top_17_LV_log
```

### Phenotype code description

```r
pheno_descr_code <- phe_T1D_LVs_nom_0.01 [,c('phenotype','phenotype_desc')]%>%
distinct()

pheno_descr_code
```

### Anova between three T1D progression groups using 17 LVs

```r
# compute anova for 17 LV
T1D_LV <- as.vector(unique(top_17_LV$lv))
res <- data.frame(matrix(nrow = 0, ncol = 6))

# compare between three groups using anova
for (i in 1:length(T1D_LV)){
    comp_i <- compare_means(as.formula(paste0(T1D_LV[i]," ~ Group")),  data = DAISY_diff_proj,
                            method = "anova")
    res <- rbind(res, comp_i)
    }
res_formated <- res[,c(1,2)]
colnames(res_formated) <- c("LV","p_val")
res_formated$fdr <- p.adjust(res_formated$p_val, method = "BH")
res_formated

#saveRDS(res_formated,paste0("output/anova_",tp,".rds"))
```

### Boxplot of three groups accross each of 17 LVs

```r
for (i in 1:length(T1D_LV)){
    plot_i <- ggboxplot(DAISY_diff_proj, x = "Group", y = T1D_LV[i], color = "Group", 
          add = "jitter", legend = "none") +
  rotate_x_text(angle = 45)+
  geom_hline(yintercept = mean(DAISY_diff_proj[,T1D_LV[i]]), linetype = 2)+ # Add horizontal line at base mean
  stat_compare_means(method = "anova", label.y = max(DAISY_diff_proj[,T1D_LV[i]])+ 0.01)       # Add global annova p-value 
    print(plot_i)
    
    #ggsave(paste0("LV_",T1D_LV[i],"_visit_",tp,".png"),path = paste0("figures/visit_",tp))
    } 
    #dev.off()
```
